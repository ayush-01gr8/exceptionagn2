```python
The ArithmeticError class in Python is the base class for exceptions that arise from arithmetic operations. It serves as a parent class for 
various specific arithmetic-related exception classes.
Two notable exceptions that inherit from ArithmeticError are ZeroDivisionError and OverflowError.
```

```python
try :
    result= 10/0
except ZeroDivisionError as e:
        print(f"caught zerodivision error : {e}")
```

```python
import sys

try:
    huge_number = sys.maxsize ** 10
except OverflowError as e:
    print(f"Caught OverflowError: {e}")
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
The LookupError class in Python is the base class for exceptions that occur when an item is not found in a collection or when an index is out of range.
It provides a common parent class for various lookup-related exceptions, making it easier to catch and handle these types of errors in a consistent 
manner.
```

```python
my_dict = {"apple": 3, "banana": 5, "orange": 2}

try:
    count = my_dict["grape"]
except KeyError as e:
    print(f"Caught KeyError: {e}")
```

```python
my_list = [1, 2, 3, 4, 5]

try:
    value = my_list[10]
except IndexError as e:
    print(f"Caught IndexError: {e}")
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?
```

```python
ImportError is an exception in Python that is raised when an import statement fails to locate and load the desired module. It typically occurs 
when Pythoncannot find the specified module in the search paths (sys.path) or when there's an issue with the module's code or dependencies.
```

```python
try:
    import non_existent_module
except ImportError as e:
    print(f"Caught ImportError: {e}")
```

```python
ModuleNotFoundError:
ModuleNotFoundError is a subclass of ImportError that specifically indicates that the module being imported could not be found.
```

```python
try:
    import non_existent_module
except ModuleNotFoundError as e:
    print(f"Caught ModuleNotFoundError: {e}")
```
